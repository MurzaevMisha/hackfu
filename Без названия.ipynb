```python
import os
import pandas as pd
import numpy as np

df = pd.read_csv('lable.csv', delimiter=';')
```

```python
for i in df.itertuples():
    i = i[1].split(',')
    name = i[0].split('.')
    f = open(f'data/train/labels/{name[0]}.txt', 'a')
    f.write(f'0 {int(i[1])/1360} {int(i[2])/800} {int(i[3])/1360} {int(i[4])/800}\n')
    f.close()
```
